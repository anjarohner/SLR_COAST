```python
import xarray as xr

# Load the NetCDF file
data = xr.open_dataset('annual_global.nc')

# Define the time span
start_year = 1984
end_year = 2015

# Extract the subset of data for the time span
subset = data.sel(time=slice(f'{start_year}-01-01', f'{end_year}-12-31'))

# Compute the mean over the time dimension
mean_data = subset.mean(dim='time')

# Save the mean data to a new NetCDF file
mean_data.to_netcdf('global_mean_sea_level_1984_2015.nc')
```
