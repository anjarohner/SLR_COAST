```python
import xarray as xr

# Open the NetCDF file using xarray
ds = xr.open_dataset('annual_global.nc')

# Extract data for the years 1984-2015
ds = ds.sel(time=slice('1984', '2015'))

# Extract the variables you want
hr_relative_cfact_no_season = ds['hr_relative_cfact_no_season']
seasonal_cycle = ds['seasonal_cycle']
lat = ds['lat']
lon = ds['lon']

# Calculate hcc_cfact by adding hr_relative_cfact_no_season and seasonal_cycle
hcc_cfact = hr_relative_cfact_no_season + seasonal_cycle

# Calculate the mean over the years 1984-2015
hcc_cfact_mean = hcc_cfact.mean(dim='time')

# Create a new dataset with the mean values and the lat/lon coordinates
ds_mean = xr.Dataset({'hcc_cfact': hcc_cfact_mean}, coords={'lat': lat, 'lon': lon})

# Save the new dataset to a NetCDF file
ds_mean.to_netcdf('hcc_cfact_mean.nc')
```
