```python
# Saving a subset of the original dataset, from January 1984 to December 2014

#ds.to_netcdf("sea_level_data_1984-2014.nc")

#lat_min = -21 lat1
#lat_max = -19 lat2
#lon_min = 55  lon1
#lon_max = 58  lon2

bbox = (-21, 55, -19, 58)

sealevel = xr.open_dataset("sea_level_data_1984-2014.nc")

coastline = xr.open_dataset('globalCoastalMorphodynamicsDb.nc')

coastline
```

```python
# Filtering an xarray Dataset based on a bounding box defined by latitude and longitude coordinates,
# returning a subset of the data

from itertools import groupby
    
def select_bbox(ds, bbox):
    lat1, lon1, lat2, lon2 = bbox
    idx = (ds.lon > lon1) & (ds.lat > lat1) & (ds.lon < lon2) & (ds.lat < lat2)
    return ds.isel(transect=idx.values, ntransect=idx.values)
```

```python
# Deriving a bounding box from countrymask.nc

coastline_bbox = select_bbox(coastline, bbox)
coastline_bbox
```

```python
# Filtering a list of station data by latitude and longitude bounds
# Defining four variables, lat_min, lat_max, lon_min, and lon_max, 
# which represent the minimum and maximum latitude and longitude values 
# station considered as "good", when in bounding box

lat_min, lon_min, lat_max, lon_max = bbox

good_stations = []

for i, station in tqdm(enumerate(sealevel["hr"].station_name)):
    lat_ok = station.lat < lat_max and station.lat > lat_min
    lon_ok = station.lon < lon_max and station.lon > lon_min
    
    if lat_ok and lon_ok:
        good_stations.append(i)
        
# All station names as list

all_station_names = [x for x in sealevel.station_name.values.ravel()] 

bad_stations = [x for i,x in enumerate(all_station_names) if i not in good_stations] 

# Remove bad stations from list
# Show the remaining data after the bad stations have been removed.

sealevel_bbox = sealevel.drop_sel(station_name=bad_stations) # Remove bad stations from data set
sealevel_bbox
```

```python
# Approximate radius of earth in km, R = 6373.0

def get_distance(lat1, lon1, lat2, lon2):
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    a = sin((lat2 - lat1) / 2)**2 + cos(lat1) * cos(lat2) * sin((lon2 - lon1) / 2)**2
    
    return 6373 * (2 * atan2(sqrt(a), sqrt(1 - a)))
```

```python
# Calculating the distances between coastline and transect points for each pair of coordinates,
# finding the minimum distance for each coastline point, and assigning the corresponding station index
# if the minimum distance is below a threshold value (100 in this case)

assignment = {}
for i, _ in tqdm(enumerate(coastline_bbox)):
    x = [] # contains the distances between the coastline and transects
        
    for j in range(len(sealevel_bbox.station_name)): 
        x.append(
            get_distance(
                sealevel_bbox.lat[j].values.ravel()[0], # lat1
                sealevel_bbox.lon[j].values.ravel()[0], # lon1
                coastline_bbox["lat"][i].values.ravel()[0], # lat2
                coastline_bbox["lon"][i].values.ravel()[0]  # lon2
            )
        )
        
        
        
        #print(station.lat)
    
    min_diff = min(x)
    
    print(x)
    
    if min_diff < 100:
        assignment[str(i)] = x.index(min_diff)
```

```python
# Extracting values from the datasets based on the assigned transect and station indices,
# calculating the difference between the mean values of the first and last 12 months for each station,
# and storing the x and y coordinates for plotting

x, y = [], [] 

transects_idxs = [int(k) for k in assignment.keys()]
station_idxs = assignment.values()

for transect_idx, station_idx  in zip(transects_idxs, station_idxs):
    x.append(coastline_bbox["activezonetoland"][transect_idx].values.ravel()[0]) # only one value
    
    year1 = sealevel_bbox["hr"][station_idx][:12].mean("time").values.ravel()[0]
    year31 = sealevel_bbox["hr"][station_idx][-12:].mean("time").values.ravel()[0]
    y.append(year1 - year31)  # A value that indicates the difference/change over all years => take the annual average for the differences
        # alternative: sealevel_bbox["hr"][station_idx].mean("time")
```
